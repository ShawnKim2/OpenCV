```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import drive
drive.mount('/content/gdrive')
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = '/content/gdrive/MyDrive/Colab Notebooks/Images/lena256.jpg'

origin_img = cv2.imread(img_path)
RGB_img = cv2.cvtColor(origin_img, cv2.COLOR_BGR2RGB)

cv2_imshow(origin_img)
```

```python
g_img = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
cv2_imshow(g_img)
```

```python
M = np.float32([[1, 0, 100], [0, 1, 50]])
height, width = g_img.shape[:2]
outImg = cv2.warpAffine(g_img, M, (width, height))
cv2_imshow(outImg)
```

```python
M = cv2.getRotationMatrix2D(((width)/2.0, (height-1)/2.0), -45, 1)
print(M)
outImg = cv2.warpAffine(g_img, M, (width, height))
cv2_imshow(outImg)
```
